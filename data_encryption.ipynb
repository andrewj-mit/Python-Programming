##Reverse Cipher
The algorithm of reverse cipher holds the following features −
* Reverse Cipher uses a pattern of reversing the string of plain text to convert as cipher text.
* The process of encryption and decryption is same.
* To decrypt cipher text, the user simply needs to reverse the cipher text to get the plain text.

```python
message = 'This program is to explain reverse cipher.'
translated = '' #cipher text is stored in this variable
i = len(message) - 1

while i >= 0:
   translated = translated + message[i]
   i = i - 1
print("The cipher text is : ", translated)
```

##Caesar Cipher
The algorithm of Caesar cipher holds the following features −
* Caesar Cipher Technique is the simple and easy method of encryption technique.
* It is simple type of substitution cipher.
* Each letter of plain text is replaced by a letter with some fixed number of positions down with alphabet.
image.png

```python
def encrypt(text,s):
  result = ""
  for i in range(len(text)):
    char = text[i] # Encrypt uppercase characters in plain text
    if (char.isupper()):
      result += chr((ord(char) + s-65) % 26 + 65)
    else:
      result += chr((ord(char) + s - 97) % 26 + 97)
  return result
#check the above function
text = "CEASER CIPHER DEMO"
s = 4

print ("Plain Text : " + text)
print ("Shift pattern : " + str(s))
print ("Cipher: " + encrypt(text,s))
```

##Generate Key
To generate a key, we call the generate_key() method:

```python
from cryptography.fernet import Fernet

"""
Generates a key and save it into a file
"""
key = Fernet.generate_key()
with open("secret.key", "wb") as key_file:
  key_file.write(key)
```

##Load the Key
Once we have generated a key, we need to load the key in our method in order to encrypt data:

```python
"""
    Loads the key named `secret.key` from the current directory.
"""
key = open("secret.key", "rb").read()
    
```

##Encrypt a Message
Now, we are ready to encrypt a message. This is a three step process:
<pre>
1 - encode the message
2 - initialize the Fernet class
3 - pass the encoded message to encrypt() method


###encode the message:

```python

message = "message I want to encrypt"
```

```python
encoded_message = message.encode()
```

###initialize the Fernet class:


```python
f = Fernet(key)
```

###encrypt the message:


```python

encrypted_message = f.encrypt(encoded_message)
```

###Print the encypted message


```python
print(encrypted_message)
```

#<b>Decrypt Data in Python</b>
To decrypt the message, we just call the decrypt() method from the Fernet library. Remember, we also need to load the key as well, because the key is needed to decrypt the message.

```python
from cryptography.fernet import Fernet

def load_key():
    """
    Load the previously generated key
    """
    return open("secret.key", "rb").read()

def decrypt_message(encrypted_message):
    """
    Decrypts an encrypted message
    """
    key = load_key()
    f = Fernet(key)
    decrypted_message = f.decrypt(encrypted_message)

    print(decrypted_message.decode())

if __name__ == "__main__":
    decrypt_message(encrypted_message)
```
